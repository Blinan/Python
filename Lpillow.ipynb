```python
import tkinter as tk
import random

class DesktopPet:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Desktop Pet")
        
        # Transparent & always-on-top setup
        self.window.overrideredirect(True)
        self.window.attributes('-topmost', True)
        self.window.wm_attributes('-transparentcolor', 'black')
        self.window.config(highlightbackground='black')
        
        # Load GIF frames (prepare your own GIFs!)
        # Example: replace with your pet GIF paths
        self.frames = [tk.PhotoImage(file='pet_idle.gif', format=f'gif -index {i}') 
                       for i in range(4)]  # Adjust frame count
        
        self.label = tk.Label(self.window, bd=0, bg='black')
        self.label.pack()
        
        self.x = 100
        self.y = 100
        self.frame_index = 0
        self.direction = 1  # 1 = right, -1 = left
        
        self.window.geometry(f'128x128+{self.x}+{self.y}')
        self.update()
        self.window.mainloop()
    
    def update(self):
        # Animate frame
        self.frame_index = (self.frame_index + 1) % len(self.frames)
        self.label.configure(image=self.frames[self.frame_index])
        
        # Simple movement
        self.x += 2 * self.direction
        if self.x > 800 or self.x < 0:  # Bounce at edges (adjust to your screen)
            self.direction *= -1
        
        self.window.geometry(f'128x128+{self.x}+{self.y}')
        
        # Schedule next update
        self.window.after(100, self.update)  # ~10 FPS

# Run the pet
if __name__ == "__main__":
    a = DesktopPet()
    while 1:
        a.update()
```
