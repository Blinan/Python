```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
csvPath = Path('data.csv')
result = pd.read_csv(csvPath)
print(result)
result.isna()
```

```python
csvPath = Path('data.csv')
result = pd.read_csv(csvPath, na_values=["NULL"])
print(result)
result.isna()
```

```python
csvPath = Path('data.csv')
result = pd.read_csv(csvPath,keep_default_na=False)
print(result)
result.isna()
```

```python
csvPath = Path('data.csv')
sentinels = {'message':['foo', 'NA'], 'something':['two']}
result = pd.read_csv(csvPath, na_values=sentinels,keep_default_na=False)
print(result)
result.isna()
```
